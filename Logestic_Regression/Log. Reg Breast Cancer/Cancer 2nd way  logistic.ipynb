```python
u = np.mean(xtrain, axis = 0)
std = np.std(xtrain, axis = 0)
```

```python
xtest = xtest.values
ytest = ytest.values
xtrain = xtrain.values
ytrain = ytrain.values
ytrain=ytrain.reshape(-1,1)
ytest=ytest.reshape(-1,1)
```

```python
def sigmoid(X):
    return 1.0/(1.0 + np.exp(-X))
```

```python
def hypothesis(X, theta):
    return sigmoid(np.dot(X, theta))
```

```python
def cost_function(X, Y, theta):
    h_theta = hypothesis(X, theta)
    cost = np.mean((Y*np.log(h_theta)) + ((1 - Y)*np.log(1 - h_theta)))
    cost = cost*-1

    return cost
```

```python
def gradient(X, Y, theta):

    h_theta = hypothesis(X, theta)

    grad = np.dot(X.T, (h_theta - Y))

    return grad/X.shape[0]
```

```python
def gradient_descent(X, Y, learning_rate = 0.001, max_steps = 500):

    m, n = X.shape
    theta = np.zeros((n, 1))
    cost_epoch = []

    for i in range(max_steps):
        print("ON: ", i)
        grad = gradient(X, Y, theta)

        e = cost_function(X, Y, theta)
        cost_epoch.append(e)

        theta = theta - learning_rate*grad

    return (theta, cost_epoch)
```

```python
theta, cost_epoch = gradient_descent(xtrain, ytrain, learning_rate=0.001, max_steps = 1000)
```

```python
def predict(X, theta):
    h_theta = hypothesis(X, theta)
    output = np.zeros(h_theta.shape)

    output[h_theta>0.5] = 1
    output = output.astype('int')

    return output
```

```python
train_preds = predict(xtrain, theta)
test_preds = predict(xtest, theta)
```

```python
for i in range(len(test_preds)):
    print(test_preds[i], ytest[i])
```

```python
def accuracy(preds, labels):
    labels = labels.astype('int')
    print(np.sum(preds == labels)/labels.shape[0])
```
